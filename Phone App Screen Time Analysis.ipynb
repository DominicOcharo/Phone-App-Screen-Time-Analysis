```python
# Phone App Screen Time Analysis using Python
# Importing python libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import plotly.express as ply
import numpy as np

path = r"C:\Users\Administrator\Documents\PythonProjects Jupyter\Apps Screen Time datasets\Screentime - App Details.csv"
data = pd.read_csv(path)
```

```python
# Observing the Dataset
def observe_data(df):
    print('OBSERVING THE NUMBER OF NULL VALUES IN THE DATASET')
    print(df.isnull().sum())
    print(' ')
    print('CHECKING THE DESCRIPTION  OF THE DATA IN THE DATASET')
    print(df.describe())
    print(' ')
    print('CHECKING THE INFORMATION ABOUT THE DATASET')
    print(df.info())
    print(' ')
    print(df.head(5))
```

```python
output = observe_data(data)
output
```

```python
# Amount of usage of the apps
def usage1(df):
    insta_usage = []
    whats_usage = []
    date = []
    for i in range(0, len(df['Date'])):
        out = df.iloc[i, 4]
        if out == 'Instagram':
            insta_usage.append(df.iloc[i, 1])
        else:
            whats_usage.append(df.iloc[i, 1])
            date.append(df.iloc[i, 0])
            
    plt.style.use('ggplot')
    plt.figure(figsize=(15,10))
    plt.bar(date, insta_usage, label='Instagram')
    plt.bar(date, whats_usage, bottom=insta_usage, label='Whatsapp')
    plt.legend(loc='best', fontsize='medium')
    plt.xticks(rotation=90)
    #plt.xlabel('date')
    #plt.ylabel('Usage Hours')
    #plt.title('Graph of Usage Hours in Apps')
    plt.show()
```

```python
output = usage1(data)
output
```

```python
# Number of notifications from the apps
def notifications(df):
    insta_notification = []
    whats_notification = []
    date = []
    for i in range(0, len(df['Date'])):
        out = df.iloc[i, 4]
        if out == 'Instagram':
            insta_notification.append(df.iloc[i, 2])
        else:
            whats_notification.append(df.iloc[i, 2])
            date.append(df.iloc[i, 0])
            
    plt.style.use('ggplot')
    plt.figure(figsize=(15,10))
    plt.bar(date, insta_notification, label='Instagram')
    plt.bar(date, whats_notification, bottom=insta_notification, label='Whatsapp')
    plt.legend(loc='best', fontsize='medium')
    plt.xticks(rotation=90)
    #plt.xlabel('date')
    #plt.ylabel('Notification Hours')
    #plt.title('Graph of Notification Hours in Apps')
    plt.show()
```

```python
output = notifications(data)
output
```

```python
# Number of times each app was opened
def app_open(df):
    insta_open_time = []
    whats_open_time = []
    date = []
    for i in range(0, len(df['Date'])):
        out = df.iloc[i, 4]
        if out == 'Instagram':
            insta_open_time.append(df.iloc[i, 3])
        else:
            whats_open_time.append(df.iloc[i, 3])
            date.append(df.iloc[i, 0])
            
    plt.style.use('ggplot')
    plt.figure(figsize=(15,10))
    plt.bar(date, insta_open_time, label='Instagram')
    plt.bar(date, whats_open_time, bottom=insta_open_time, label='Whatsapp')
    plt.legend(loc='best', fontsize='medium')
    plt.xticks(rotation=90)
    #plt.xlabel('date')
    #plt.ylabel('Times Opened')
    #plt.title('Apps Opening Times')
    plt.show()
```

```python
output = app_open(data)
output
```

```python
# Relationship between the number of notifications and the amount of usage
def usage_vs_notifications1(df):
    usage = df['Usage']
    notifications = df['Notifications']
    plt.style.use('ggplot')
    plt.figure(figsize=(15, 10))
    plt.scatter(notifications, usage, c=notifications, s=notifications, cmap='summer')
    cbar = plt.colorbar()
    cbar.set_label("Notification level")
    z = np.polyfit(notifications, usage, 1)
    p = np.poly1d(z)
    plt.plot(notifications, p(notifications), c='green', linestyle='--')
    plt.xlabel='Notifications'
    plt.ylabel='Usage'
    plt.title='Relationship between the number of notifications and the amount of usage'
    plt.show()
```

```python
output = usage_vs_notifications1(data)
output
```
